```python
import pandas as pd
import numpy as np
from datetime import date
from jugaad_data.nse import bhavcopy_save, bhavcopy_fo_save
from jugaad_data.nse import stock_df
import pandas_ta as ta
```

```python
df = stock_df(symbol="SBIN", from_date=date(2023,1,1),
            to_date=date(2023,10,1), series="EQ")
```

```python
df
```

```python
ndf = pd.DataFrame()
```

```python
rsi_period = 14  # Set your desired RSI period
cci_period = 14  # Set your desired CCI period
adx_period = 14  # Set your desired ADX period

```

```python
ndf['RSI'] = ta.rsi(df['CLOSE'], length=rsi_period)
ndf['CCI'] = ta.cci(df['HIGH'], df['LOW'], df['CLOSE'], length=cci_period)
```

```python
adx_df = ta.adx(df['HIGH'], df['LOW'], df['CLOSE'], length=adx_period)
ndf = ndf.join(adx_df,how='inner')
```

```python
short_sma = df['CLOSE'].rolling(window=5).mean()
long_sma = df['CLOSE'].rolling(window=34).mean()
awesome_oscillator = short_sma - long_sma
```

```python
wt = awesome_oscillator * ndf['RSI']
```

```python
wta_period = 9
wta = wt.rolling(window=wta_period).mean()
```

```python
ndf['WT'] = wt
ndf['WTA'] = wta
```
